# Missing Data

```python
import pandas as pd

df =  pd.DataFrame({'A': [1, 2, np.nan, 4],  
                  'B': [5, np.nan, np.nan, 8],
                  'C': [10, 20, 30, 40 ]})

df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh = 2)
```

```python
df.fillna(value = 'Fill Value')
```

```python
df['A'].fillna(value = 0)
```

```python
df['A'].fillna(df['A'].mean())
```

```python
df.fillna(df.mean())
```

# Groupby

```python
import pandas as pd
```

```python
df = pd.read_csv('Universities.csv')
```

```python
df.head()
```

```python
# Step 1 simply returns a special groupby object waiting to have an aggregate method called on it!
df.groupby('Year')
```

```python
df.groupby('Year').mean()
```

```python
type(df.groupby('Year').mean())
```

```python
df.groupby('Year').mean().sort_index(ascending=False)
```

```python
df.groupby('Year').mad()
```

```python
df.groupby('Year').median()
```

```python
df.groupby('Year').std()
```

```python
df.groupby('Year').skew()
```

```python
df.groupby('Year').var()
```

```python
df.head()
```

```python
df.groupby(['Year', 'Sector']).mean()
```

```python
df.groupby('Year').describe()
```

```python
df.groupby('Year').describe().transpose()
```

```python
df_one
```

```python
df_one['col2'].unique()
```

```python
df_one['col2'].nunique()
```

```python
df_one['col2'].value_counts()
```

```python
df_one.drop_duplicates()
```

```python
df_one['New Col'] = df_one['col1'] * 10
df_one
```

```python
def grab_first_letter(state):
    return state[0]
```

```python
grab_first_letter('NY')
```

```python
df_one['col2'].apply(grab_first_letter)
```
